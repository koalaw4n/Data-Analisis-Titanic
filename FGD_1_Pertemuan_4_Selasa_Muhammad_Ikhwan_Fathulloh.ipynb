```python
f,axes=plt.subplots(ncols=2,nrows=4,figsize=(10,10),sharex=True)
l=['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp',
       'Parch', 'Fare']
     
for i in range(4):
    
    sns.boxplot(x = titanic.loc[:,l[i]], ax=axes[i][0], color='powderblue')
   
    sns.histplot(data=titanic.loc[:,l[i]], color="red", kde=True, stat="density", linewidth=0, ax=axes[i][1])  
    
    plt.tight_layout()
```

```python
titanic['Sex'] = titanic.Sex.replace({'male': 0, 'female' : 1})
titanic
```

```python
sns.pairplot(titanic, hue="Survived", diag_kind="hist")
plt.show()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=titanic['Survived'], y=titanic['Age'], ax=axes[0])

sns.boxplot(x=titanic['Survived'], y=titanic['Age'], ax=axes[1])

plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=titanic['Survived'], y=titanic['Sex'], ax=axes[0])

sns.boxplot(x=titanic['Survived'], y=titanic['Sex'], ax=axes[1])

plt.tight_layout()
```

```python
f, axes = plt.subplots(ncols = 2, figsize=(10, 5), sharex=True)
sns.despine(left=True)

sns.swarmplot(x=titanic['Survived'], y=titanic['Embarked'], ax=axes[0] ,size=4.5)

sns.boxplot(x=titanic['Survived'], y=titanic['Embarked'], ax=axes[1])

plt.tight_layout()
```
